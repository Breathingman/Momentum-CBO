```python
from matplotlib import pyplot as plt

f = open("Results\\1.Strict\\Sink\\consensus.txt","r")
consensus = []
i = 0
while True:
    line = f.readline()
    if not line: break
    line = line.replace("Step "+str(i)+": tensor(","")
    line = line.replace(", dtype=torch.float64)\n","")
    i = i+1
    consensus.append(float(line))

plt.figure()
plt.plot(consensus)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

f1 = open("Results\\ADAM_CBO\\2.Invariant\\Pendulum_7\\losses.txt","r")
f2 = open("Results\\CBO\\2.Invariant\\Pendulum_1\\losses.txt","r")
f3= open("Results\\GD\\2.Invariant\\Pendulum_5\\losses.txt","r")
f4= open("Results\\GD\\2.Invariant\\Pendulum_3\\losses.txt","r")
f5= open("Results\\Momentum_CBO\\2.Invariant\\Pendulum_5\\losses.txt","r")
f6 = open("Results\\SGD\\2.Invariant\\Pendulum_4\\losses.txt","r")
f7 = open("Results\\Adam\\2.Invariant\\Pendulum_1\\losses.txt","r")
consensus1 = []
consensus2 = []
consensus3 = []
consensus4 = []
consensus5 = []
consensus6 = []
consensus7 = []
i = 0
while True:
    line = f1.readline()
    if not line: break

    line = line.split(" ")[2]
    line = line.replace("Step ","")
    line = line.replace("tensor(","")
    line = line.replace(",","")
    consensus1.append(math.log(abs(float(line))+1e-12))

while True:
    line = f2.readline()
    if not line: break

    line = line.split(" ")[2]
    line = line.replace("Step ","")
    line = line.replace("tensor(","")
    line = line.replace(",","")
    consensus2.append(math.log(abs(float(line))+1e-12))

while True:
    line = f3.readline()
    if not line: break

    line = line.split(" ")[2]
    line = line.replace("Step ","")
    line = line.replace("tensor(","")
    line = line.replace(",","")
    consensus3.append(math.log(abs(float(line))+1e-12))

while True:
    line = f4.readline()
    if not line: break

    line = line.split(" ")[2]
    line = line.replace("Step ","")
    line = line.replace("tensor(","")
    line = line.replace(",","")
    consensus4.append(math.log(abs(float(line))+1e-12))

while True:
    line = f5.readline()
    if not line: break

    line = line.split(" ")[2]
    line = line.replace("Step ","")
    line = line.replace("tensor(","")
    line = line.replace(",","")
    consensus5.append(math.log(abs(float(line))+1e-12))

while True:
    line = f6.readline()
    if not line: break

    line = line.split(" ")[2]
    line = line.replace("Step ","")
    line = line.replace("tensor(","")
    line = line.replace(",","")
    consensus6.append(math.log(abs(float(line))+1e-12))

while True:
    line = f7.readline()
    if not line: break

    line = line.split(" ")[2]
    line = line.replace("Step ","")
    line = line.replace("tensor(","")
    line = line.replace(",","")
    consensus7.append(math.log(abs(float(line))+1e-12))

plt.figure()
plt.xlabel("iteration")
plt.ylabel("loss_logscale")
plt.plot(consensus6, label = 'SGD_Failure')
plt.plot(consensus3, label = 'GD_Failure')
plt.plot(consensus7, label = "Adam_Failure", color = 'k')
plt.plot(consensus4, label = 'GD_Success')
plt.plot(consensus1, label = 'Adam-CBO')
plt.plot(consensus2, label = 'CBO')
plt.plot(consensus5, label = 'Momentum-CBO')
plt.legend()
plt.show()
```
